## 13.1 运算符重载基础

* 规则：
    - 不能重载内置类型的运算符；
    - 不能新建运算符，只能重载现有的；
    - 某些运算符不能重载,(is, and, or, not); 位运算符(&, |, ~)可以重载。

* 一元运算符：
    - `- (__neg__)`：取负；
    - `+ (__pos__)`：取正运算符；
    - `~ (__invert__)`: 按位取反；
    - `abs (__abs__)`: 求绝对值；
* 中缀运算符：(反向方法，就地方法)


* 比较运算符：



- 精度变化 `x != +x`

```python
from decimal import Decimal, getcontext

getcontext().prec = 40   # 保留40位精度

one_third = Decimal('1') / Decimal('3')
print(f"one_third \t\t\t {one_third}")
print(f"one_third == +one_third \t {one_third == +one_third}")

getcontext().prec = 30   # 保留30位精度
print(f"one_third == +one_third \t {one_third == +one_third}")
print(f"+one_third \t\t\t {+one_third}")  # 创建新的精度实例
```

```python
from collections import Counter

ct = Counter('abrasdfghasdfgwdfvbnazf')
print(f"ct \t {ct}")

ct['r'] = -3
print(f"ct \t {ct}")    # 按序输出
print(f"+ct \t {+ct}")  # 按序取正值输出
```
